## Dummy Variables


```python
cars_dummy = pd.get_dummies(Cars['fuelsystem'], prefix='fuelsystem')
```

```python
Cars = pd.concat([Cars, cars_dummy], axis=1)
```

```python
Cars = Cars.drop('fuelsystem', axis=1)
```

```python
Cars.head()
```

```python
engine_dummy = pd.get_dummies(Cars['enginetype'], prefix='enginetype')
```

```python
Cars = pd.concat([Cars, engine_dummy], axis=1)
```

```python
Cars = Cars.drop('enginetype', axis=1)
```

```python
Cars = Cars.drop('CarName', axis = 1)
```

```python
Cars.head()
```

```python
def convert_number(x):
    door_dict = { 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8 , 'twelve': 12}
    return door_dict.get(x)
```

```python
Cars['doornumber'] = Cars['doornumber'].apply(convert_number)
```

```python
Cars.head()
```

```python
fuel_dummy = pd.get_dummies(Cars['fueltype'], prefix='fueltype')
```

```python
Cars = pd.concat([Cars, fuel_dummy], axis=1)
```

```python
Cars = Cars.drop('fueltype', axis=1)
```

```python
Cars.head()
```

```python
aspiration_dummy = pd.get_dummies(Cars['aspiration'], prefix='aspiration')
Cars = pd.concat([Cars, aspiration_dummy], axis=1)
Cars = Cars.drop('aspiration', axis=1)
```

```python
carbody_dummy = pd.get_dummies(Cars['carbody'], prefix='carbody')
Cars = pd.concat([Cars, carbody_dummy], axis=1)
Cars = Cars.drop('carbody', axis=1)
```

```python
Cars.head()
```

```python
drivewheel_dummy = pd.get_dummies(Cars['drivewheel'], prefix='drivewheel')
Cars = pd.concat([Cars, drivewheel_dummy], axis=1)
Cars = Cars.drop('drivewheel', axis=1)
```

```python
enginelocation_dummy = pd.get_dummies(Cars['enginelocation'], prefix='enginelocation')
Cars = pd.concat([Cars, enginelocation_dummy], axis=1)
Cars = Cars.drop('enginelocation', axis=1)
```

```python
CarBrand_dummy = pd.get_dummies(Cars['CarBrand'], prefix='CarBrand')
Cars = pd.concat([Cars, CarBrand_dummy], axis=1)
Cars = Cars.drop('CarBrand', axis=1)
```

```python
ModelName_dummy = pd.get_dummies(Cars['ModelName'], prefix='ModelName')
Cars = pd.concat([Cars, ModelName_dummy], axis=1)
Cars = Cars.drop('ModelName', axis=1)
```

```python
def change(x):
    door_dict1 = { 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8 , 'twelve': 12}
    return door_dict1.get(x)
```

```python
Cars['cylindernumber'] = Cars['cylindernumber'].apply(change)
```

```python
Cars.head()
```

We have converted all categorical vars into numerical variables

```python
from sklearn.model_selection import train_test_split


np.random.seed(0)
df_train, df_test = train_test_split(Cars, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight','enginesize','boreratio','stroke','compressionratio','horsepower',
            'peakrpm','citympg','highwaympg']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()  
```

```python
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_new = X_train_rfe.drop(["enginelocation_front"], axis = 1)
```

```python
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_lm).fit() 
```

```python
print(lm.summary())
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_final = X_train_new.drop('const', axis = 1)
```

```python
lm_final = sm.OLS(y_train,X_train_final).fit() 
```

```python
print(lm.summary())
```

```python
y_train_price = lm.predict(X_train_lm)
```
